```python
import gensim
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir('/kaggle/input/harry-potter-books'):
    
    f = open(os.path.join('/kaggle/input/harry-potter-books',filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('voldemort')
```

```python
model.wv.doesnt_match(['harry','voldemort','dumbledore','hogwarts','weasley'])
```

```python
model.wv.doesnt_match(['harry', 'dumbledore', 'hogwarts', 'weasley'])
```
